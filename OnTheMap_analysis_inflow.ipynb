```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
fname = "https://raw.githubusercontent.com/ReidelVichot/DE_ASCENT/main/OnTheMap_inflow.csv"
df = pd.read_csv(fname, parse_dates=["year"])
```

```python
count = df[df["type"] == "Count"]
share = df[df["type"] == "Share"]
```

```python
df.columns
```

```python
df.dtypes
```

```python
count["Net Job Inflow (+) or Outflow (-)"] = count["Net Job Inflow (+) or Outflow (-)"].astype(int)
```

```python
|count.columns
```

```python
fig, ax = plt.subplots(figsize= (10,5))
count.plot(y='External Jobs Filled by Residents', x="year", ax=ax)
count.plot(y='Internal Jobs Filled by Outside Workers', x="year", ax=ax)
count.plot(y='Internal Jobs Filled by Residents', x="year", ax=ax)
ax.set_title("Count of Job Flow")
fig.show()
```

```python
fig, ax = plt.subplots(figsize= (10,5))
count.plot(y='Employed in the Selection Area', x="year", ax=ax)
count.plot(y='Living in the Selection Area', x="year", ax=ax)
count.plot(y='Net Job Inflow (+) or Outflow (-)', x="year", ax=ax)
ax.set_title("Employed vs Living")
fig.show()
```

```python
fig, ax = plt.subplots(figsize= (10,5))
share.plot(y='Employed in the Selection Area', x="year", ax=ax)
share.plot(y='Living in the Selection Area', x="year", ax=ax)
ax.set_title("Employed vs Living (Share)")
fig.show()
```

```python
fig, ax = plt.subplots(1,2, figsize= (20,5))

ax[0].set_title("In-Area Labor Force Efficiency (All Jobs)")
count.plot(y="Living in the Selection Area", x="year", ax=ax[0])
count.plot(y="Living and Employed in the Selection Area", x="year", ax=ax[0])
count.plot(y="Living in the Selection Area but Employed Outside", x="year", ax=ax[0])

ax[1].set_title("In-Area Employment Efficiency (All Jobs)")
count.plot(y="Employed in the Selection Area", x="year", ax=ax[1])
count.plot(y="Employed and Living in the Selection Area", x="year", ax=ax[1])
count.plot(y="Employed in the Selection Area but Living Outside", x="year", ax=ax[1])

ax[0].set_xlabel("")
ax[0].set_ylabel("Number of Employees")
ax[1].set_xlabel("")

fig.show()
```

```python
fig, ax = plt.subplots(1,2, figsize= (20,5))

ax[0].set_title("In-Area Labor Force Efficiency (All Jobs)")
share.plot(y="Living in the Selection Area", x="year", ax=ax[0])
share.plot(y="Living and Employed in the Selection Area", x="year", ax=ax[0])
share.plot(y="Living in the Selection Area but Employed Outside", x="year", ax=ax[0])

ax[1].set_title("In-Area Employment Efficiency (All Jobs)")
share.plot(y="Employed in the Selection Area", x="year", ax=ax[1])
share.plot(y="Employed and Living in the Selection Area", x="year", ax=ax[1])
share.plot(y="Employed in the Selection Area but Living Outside", x="year", ax=ax[1])

ax[0].set_xlabel("")
ax[0].set_ylabel("Share of Employees")
ax[1].set_xlabel("")

fig.show()
```
